# Step 1: Inventory Data Exploration

Load and explore the dataset to understand:
- Inventory levels
- Units sold (demand)
- Demand forecast
- Pricing & discounts
- Seasonality & promotions
- Store & region variations

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path('..').resolve()))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.data_exploration import load_data, explore_dataset
```

```python
# Load data (create sample if needed)
data_path = Path('../data/inventory_data.csv')
if not data_path.exists():
    from data.create_sample_data import create_sample_dataset
    create_sample_dataset(5000, str(data_path))

df = load_data(str(data_path))
df.head(10)
```

```python
# Summary statistics
summary = explore_dataset(df)
print('Shape:', summary['shape'])
print('\nInventory:', summary['inventory'])
print('\nUnits sold:', summary['units_sold'])
print('\nDemand variability (CV):', summary['demand_variability'])
```

```python
# Visualize demand distribution
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

df['units_sold'].hist(ax=axes[0,0], bins=50, edgecolor='black', alpha=0.7)
axes[0,0].set_title('Units Sold Distribution')

df['inventory_level'].hist(ax=axes[0,1], bins=50, edgecolor='black', alpha=0.7)
axes[0,1].set_title('Inventory Level Distribution')

df.groupby('region')['units_sold'].mean().plot(kind='bar', ax=axes[1,0])
axes[1,0].set_title('Avg Demand by Region')
axes[1,0].tick_params(axis='x', rotation=45)

df.groupby('is_promotion')['units_sold'].mean().plot(kind='bar', ax=axes[1,1])
axes[1,1].set_title('Avg Demand: Promotion vs No Promotion')
axes[1,1].set_xticklabels(['No Promo', 'Promo'], rotation=0)

plt.tight_layout()
plt.show()
```

```python
# Seasonality and price correlation
if 'date' in df.columns:
    df['month'] = df['date'].dt.month
    monthly_demand = df.groupby('month')['units_sold'].mean()
    monthly_demand.plot(kind='line', marker='o', figsize=(10, 4))
    plt.title('Monthly Demand Pattern (Seasonality)')
    plt.xlabel('Month')
    plt.ylabel('Avg Units Sold')
    plt.show()
```
